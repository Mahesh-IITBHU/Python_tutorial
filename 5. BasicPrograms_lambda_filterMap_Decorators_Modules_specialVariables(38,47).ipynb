# Problem 1 : Fibonacci Sequence

- pattern : 0 1 1 2 3 5 8 13 21........ (next digit is the sum of last 2 consecutive digit)

```python
def fib(n):
    
    a = 0
    b = 1
       
    if n < 0:
        print('wrong input')
               
    elif n >= 1:
        print(a,b)
    
    for i in range(2,n):
        c = a + b                        #sequence reqd to add
        a = b                            #swap concept to move iterator
        b = c
        print(c, end = " ")
```

```python
fib(19)
```

```python
# Program to display the Fibonacci sequence up to n-th term

nterms = int(input("How many terms? "))

# first two terms
n1, n2 = 0, 1
count = 0

# check if the number of terms is valid
if nterms <= 0:
   print("Please enter a positive integer")
# if there is only one term, return n1
elif nterms == 1:
   print("Fibonacci sequence upto",nterms,"terms:")
   print(n1)
    
# generate fibonacci sequence
else:
   print("fibonacci sequence upto",nterms,"terms : ")
   while count < nterms:
       print(n1, end = " ")
       nth = n1 + n2
       # update values
       n1 = n2
       n2 = nth
       count += 1
```

```python
def fibSeq(nterms):   
    
    n1, n2 = 0, 1
    count = 0
    if nterms <= 0:
        print("Please enter a positive integer")

    elif nterms == 1:
        print("Fibonacci sequence upto",nterms,":")
        print(n1)
    
    else:
        print("fibonacci sequence upto",nterms,"terms is : ")
        while count < nterms:
            if count == nterms:
                break
            print(n1, end = " ")
            nth = n1 + n2
            n1 = n2
            n2 = nth
            count += 1     

nterms = int(input("How many terms? "))
fibSeq(nterms)
```

```python
fibSeq(9)
```

```python
# Python program to display the Fibonacci sequence

def recur_fibo(n):
   if n <= 1:
       return n
   else:
       return(recur_fibo(n-1) + recur_fibo(n-2))

nterms = int(input("Fibonacci sequence upto: "))

# check if the number of terms is valid
if nterms <= 0:
   print("Plese enter a positive integer")
else:
   print("Fibonacci sequence upto",nterms,"")
   for i in range(nterms):
        print(recur_fibo(i), end = " ")                   #calling function recursively 
```

# Factorial Calculation

- 5! = 5 x 4 x 3 x 2 x 1
- 9! = 9 x 8 x 7 x 6 x 5 x 4 x 3 x 2 x 1 etc.


We can either go in ascending order i.e 1x2x3x4x5 OR  desc order 5x4x3x2x1 etc

```python
import math

n = int(input("Enter number for factorial calculation: "))
print(math.factorial(n))
```

```python
def fact(n):
    f = 1
    for i in range(1,n+1):                                    #start from 1 not from 0 and end at n hence (n+1)    
        f*=i
    return f
        
        
n = int(input("Enter number for factorial calculation: "))
fact(n)
```

```python
def fact(n):
    f = 1
    
    if n < 0:
        print("factorial doesn't exist for negative numbers")
    elif n == 0:
        print('Factorial of 0 is 1')
    else:
        for i in range(1,n+1):                                    
            f*=i
        return f
        
n = int(input("Enter number for factorial calculation: "))
fact(n)
```

```python
# Factorial Calculation using Recursion:-

def fact(n):
    if n==0:
        return 1
    
    return n * fact(n-1)

n = int(input("Enter number for factorial calculation: "))
print(fact(n))
```

# Filter

```python
# Filter takes 2 arguments :  function, iterable

def is_even(n):
    return n%2==0

nums = [4,3,7,6,9,12,18,15,16,19,23,22,26,33]
even = list(filter(is_even,nums))         #filter out even no using is_even function and store in the list of even.
print(even)


# function are created so that can be used multiple time as required
# here we are using is_even function customised only to be used for filter method, we can use lambda function instead.
```

```python
nums = [4,3,7,6,9,12,18,15,16,19,23,22,26,33]
even = list(filter(lambda n : n%2==0,nums))         
print(even)
```

# Map

```python
# Map : Help in applying operation on required data only after filter

def update(n):
    return n*2

nums = [4,3,7,6,9,12,18,15,16,19,23,22,26,33]
even = list(filter(lambda n : n%2==0,nums))

doubles = list(map(update,even))
print(doubles)
```

```python
nums = [4,3,7,6,9,12,18,15,16,19,23,22,26,33]
even = list(filter(lambda n : n%2==0,nums))
doubles = list(map(lambda n : n*2,even))
print(doubles)
print(even)
```

# Reduce

#### Reduce Function

- Generally when we add , we add 2 values at a time for e.g  [2,5,6,8,9,10] = 2+5;7+6;13+8;21+9;30+10 = 40
- But to perform addition at once like  2+5+6+8+9+10

To use **reduce** function import it from functools
> Reduce takes 2 arguments = function,sequence


```python
from functools import reduce
```

```python
def add_all(a,b):
    return a+b

nums = [4,3,7,6,9,12,18,15,16,19,23,22,26,33]
even = list(filter(lambda n : n%2==0,nums))
doubles = list(map(lambda n : n*2,even))

sum = reduce(add_all,doubles)
print(sum)
```

```python
nums = [3,2,6,8,4,6,2,9]
even = list(filter(lambda n : n%2==0,nums))
doubles = list(map(lambda n : n*2,even))
print(even)
print(doubles)

sum = reduce(lambda a,b : a + b,doubles)
print(sum)
```

# Decorators

> - ### Using decorators we can add extra features in the existing functions at the compile time itself+
> - ### Python Decorators make an extensive use of closures

```python
def div(a,b):
    print(a/b)
```

```python
div(2,4)
div(4,2)
```

```python
# what if it require to always divide such that numerator > denominator:
def div(a,b):
    print(a/b)
    
def smart_div(func):
    
    def inner(a,b):           #inner func takes same parameter as that of div
        if a<b:
            a,b = b,a
        return func(a,b)
    return inner

div = smart_div(div)          # creating connection between smart_div and div
div(2,4)
```

## Decorators can solve following issue:

1. Code duplication
2. Cluttering main logic of function with additional functionality (i.e timing in our example)
3. Decorators act as a wrapper for the original function

### Function : 

> Function are first class objects in python, meaning that they can be treated just like any other variable and we can pass them as argument to  another function or even return them as a return value.

##  *args     = positional argument
## *kwargs = keyword argument

```python
import time

def calc_square(nums):
    
    start = time.time()
    result = []
    for n in nums:
        result.append(n*n)
    end = time.time()
    print('calc_square took ' + str((end-start)*1000) + ' mil sec')
    return result

def calc_cube(nums):
    
    start = time.time()
    result = []
    for n in nums:
        result.append(n*n*n)
    end = time.time()
    print('calc_square took ' + str((end-start)*1000) + ' mil sec')
    return result

array = range(1,100000)
out_square = calc_square(array)
out_cube = calc_cube(array)
```

```python
import time

def time_it(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func (*args, **kwargs)
        end = time.time()
        print (func.__name__ + " took " + str((end - start) * 1000) + " mil sec")
        return result       
    return wrapper
               
@time_it              
def calc_square(nums):   
    result = []
    for n in nums:
        result.append(n*n)
    return result

@time_it
def calc_cube(nums):
    result = []
    for n in nums:
        result.append(n*n*n)
    return result

array = range(1,100000)
out_square = calc_square(array)
out_cube = calc_cube(array)
```
